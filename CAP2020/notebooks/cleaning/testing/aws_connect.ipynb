### UPLOAD FILE TO AWS

```python
data_path = 'C:\\Users\\A Sua\\Documents\\FIU\\CAP2020\\data\\external'
file_name = 'old_key_new_key.csv'
file_path = data_path + '\\' + file_name


BUCKET_NAME = 'capstonefiu2020-data'
key_location = 'data/external'
key_path = key_location +'/'+file_name


## FROM capstoner user rol
data = open(file_path, 'rb')

s3 = boto3.resource('s3', 
              aws_access_key_id = AWS_ACCESS_KEY_ID,
               aws_secret_access_key=AWS_SECRET_ACCESS_KEY)

s3.Bucket(BUCKET_NAME).put_object(Key=key_path, Body=data)

data.close()
```

### READ DATAFRAME FROM S3 BUCKET

```python
client = boto3.client('s3', 
              aws_access_key_id = AWS_ACCESS_KEY_ID,
               aws_secret_access_key=AWS_SECRET_ACCESS_KEY)
```

```python
csv_obj = client.get_object(Bucket = BUCKET_NAME, Key='data/interim/df_askhi.csv')
body = csv_obj['Body']
csv_string = body.read().decode('utf-8')

df = pd.read_csv(StringIO(csv_string))
df.index = df['Unnamed: 0']
df = df.drop('Unnamed: 0', axis=1)
df.index.names = ['key']
df.head()
```
