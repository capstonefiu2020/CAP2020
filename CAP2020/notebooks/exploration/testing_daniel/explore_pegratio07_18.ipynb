## Consolidate Indices

```python
idx1 = df_price.index
idx2 = df_eps.index
idx3 = df_div.index

idx_diff1=idx1.difference(idx2)
idx_diff2=idx2.difference(idx1)
idx_diff3=idx1.difference(idx3)
idx_diff4=idx3.difference(idx1)
idx_diff5=idx2.difference(idx3)
idx_diff6=idx3.difference(idx2)
```

```python
idx_diff2
```

```python
idx_drop_price = idx_diff1.union(idx_diff3)
df_price.drop(idx_drop_price, inplace=True)
df_price.shape
```

```python
idx_drop_eps = idx_diff2.union(idx_diff5)
df_eps.drop(idx_drop_eps, inplace=True)
df_eps.shape
```

```python
idx_drop_div = idx_diff4.union(idx_diff6)
df_div.drop(idx_drop_div, inplace=True)
df_div.shape
```

```python
df_growth.drop(idx_drop_eps, inplace=True)
df_growth.shape
```
